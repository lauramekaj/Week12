# R Code

```r
library(expm)
```

```r
x <- matrix(c(-3, 3, 5, -5), byrow=TRUE, nrow=2)
```

```r
x
```

```r
exp_m <- expm(x*3)
exp_m
```

```r
t <- seq(0, 1, by = .01)
m <- array(0, dim=c(length(t),2,2))
```

```r
for (i in 1:length(t)){
    m[i,,] <- expm(x*t[i])
}
```

```r
plot(t,m[,1,1], type = "l", ylim = c(0,1),col = "blue")
lines(t,m[,1,2], type = "l", col = "orange")
lines(t,m[,2,1], type = "l", col = "red")
lines(t,m[,2,2], type = "l", col = "green")
```

```r
from scipy.linalg import expm
import numpy as np
import matplotlib.pyplot as plt
```

```r
x = np.array([[-3,3],[5,-5]])
```

```r
expm(x*t[0])
```

```r
t = np.arange(0, 1, 0.01)
m = np.zeros((len(t),2,2))
m1_1= np.zeros(len(t))
m1_2= np.zeros(len(t))
m2_1= np.zeros(len(t))
m2_2= np.zeros(len(t))
```

```r
for i in range(0, len(t)):
    m[i] = expm(x*t[i])
    m1_1[i] = m[i,0,0]
    m1_2[i] = m[i,0,1]
    m2_1[i] = m[i,1,0]
    m2_2[i] = m[i,1,1]
```

```r
plt.plot (t,m1_1)
plt.plot (t,m1_2)
plt.plot (t,m2_1)
plt.plot (t,m2_2)
plt.show()
```
